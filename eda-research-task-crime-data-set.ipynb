```python
#Conversion of column into rows to plot relationship between Crime count VS State  

df = cy.set_index(['State/UT','Mid-Year Projected Population (in Lakhs) (2018)+','Rate of Total Cyber Crimes (2018)++','State', 'Union Territory'])\
.stack()\
.reset_index(name='CrimeCount')\
.rename(columns={'level_5':'Year'})
```

```python
df = df.drop(columns=['Mid-Year Projected Population (in Lakhs) (2018)+','Rate of Total Cyber Crimes (2018)++','State', 'Union Territory'])
```

```python
df.head(10)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'cy' DataFrame is already loaded

# Locate the outliers based on the '2018' column
Q1 = cy['2018'].quantile(0.35)
Q3 = cy['2018'].quantile(0.65)
IQR = Q3 - Q1
LowerFence = Q1 - 1.5 * IQR
HigherFence = Q3 + 1.5 * IQR

outliers = cy[cy['2018'] > HigherFence]

# Select top 7 states with maximum crime rates
top_states = outliers.nlargest(10, 'Rate of Total Cyber Crimes (2018)++')

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_states['Rate of Total Cyber Crimes (2018)++'], labels=top_states['State/UT'], autopct='%1.1f%%', startangle=140)
plt.title('Top 7 States with Maximum Crime Rates (2018)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'cy' DataFrame is already loaded

# Identify the top 15 states with higher crime rates
top_states = cy.nlargest(15, 'Rate of Total Cyber Crimes (2018)++')

# Plot a pie chart for the selected column ('Rate of Total Cyber Crimes (2018)++')
plt.figure(figsize=(10, 10))
plt.pie(top_states['Rate of Total Cyber Crimes (2018)++'], labels=top_states['State/UT'], autopct='%1.1f%%', startangle=140, colors=sns.light_palette("skyblue", n_colors=15))
plt.title(' States with Higher Crime Rates (2018)')
plt.show()
```
